## First Explorations

Provide an example of a mixed-signal circuit
- Something like a motor drive might be a mixed signal circuit, with analog circuitry for things like sensing, while PWM control for power and something like field oriented control would be implemented digitally

What type of quantities are typically measured for the control and protection of power converters?
- Current, Voltage, Temperature

What is the "loop" in loop shaping?
- The transfer function of a closed loop system

How does mechanical engineering apply to the design of power converters?
- Mechanical engineering primarily comes into play in the packaging phase, whether that's the actual assembly of the components and/or how that affects the system thermals, as well as any other requirements of the system (EMI, operating environment, space allowance, etc.)

What purpose do inductors serve in power converters?
- Energy storage, primarily

What is currently "all the rage" in the field of power electronics?
- Wide-bandgap semiconductors have created a lot of buzz due to their ability to operate in a wider temperature range and exhibit fewer switching losses than traditional semiconductors

Who is James Worden?
- James Worden is the cofounder of Solectria, who said his number one metric for assessing a power converter was efficiency


Why does the efficiency of a power converter matter?
- Higher efficiency is associated with more reliable systems, reduced thermal stress, and of course fewer energy losses, which all tend to be desirable outcomes 


Which are some of the key powerstage components?
- Components of the powerstage include power semiconductors, magnetic components, capacitors, and protective elements.


List multiple ways of getting hurt or killed when working on a power converter.
- Struck by debris from a blown component
- Electric shock
- Burns

How do you plan on staying safe.
- Observing recommended safety precautions when working with lab equipment, including
    - Not touching powered hardware
    - Wearing safety glasses and appropriate PPE
    - Operating the hardware within recommended voltage/power ranges
    - Operating converters with a resistive load attached to the output
